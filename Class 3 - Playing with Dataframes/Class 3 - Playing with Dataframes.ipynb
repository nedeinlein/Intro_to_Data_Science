```python
def filterreport(position, age, leagues, price, playerskill, df):
    df['playerskill_bin'] = pd.cut(df['overall'], bins=range(0, 101, 10), right=False)
    df['playerskill_bin'] = df['playerskill_bin'].astype(str)
    df = df[df['player_positions'].str.contains(position)]
    df = df[(df['age'] > age[0])]
    df = df[(df['age'] < age[1])]
    df = df.loc[df['league_name'].isin(leagues)]
    df = df[(df['wage_eur'] < price)]
    df = df[df['playerskill_bin']==playerskill]
    return df
```

```python
position = 'ST'
age = [17,22]
leagues = ['Premier League']
price = 5000
playerskill = '[70, 80)'

filtered = filterreport(position, age, leagues, price,playerskill, df)
filtered.head()
```
