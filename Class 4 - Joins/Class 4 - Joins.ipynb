```python
#import packages
import pandas as pd
import numpy as np
```

## import tables

```python
league_table = pd.read_csv("D:/Downloads/Class 4 - Joins/league_table.csv")
league_table.head()
```

```python
player_table = pd.read_csv("D:/Downloads/Class 4 - Joins/player_table.csv")
player_table.head()
```

```python
team_table = pd.read_csv("D:/Downloads/Class 4 - Joins/team_table.csv")
team_table.head()
```

## Make columns named the same thing

```python
team_table = team_table.rename(columns={'club_team_id':'team_id'})
team_table.head()
```

## joins

```python
#left
df = team_table.merge(league_table, on = ['league_id'], how = 'left')
df.head()
```

```python
#inner
df = team_table.merge(league_table, on = ['league_id'], how = 'inner')
df.head()
```

```python
#outer
df = team_table.merge(league_table, on = ['league_id'], how = 'outer')
df.head(100)
```

## Removing duplications

```python
team_table = team_table.drop_duplicates(subset = ['team_id'], keep = 'last')
team_table.head()
```

```python
df = team_table.merge(league_table, on = ['league_id'], how = 'left')
df.head()
```

```python
league_table = league_table.drop_duplicates(subset = ['league_id'], keep = 'last')

df = team_table.merge(league_table, on = ['league_id'], how = 'left')
df.head()
```
